```python
# dependencies
import pandas as pd
```

Cleaning Data

Clean Salary Data

```python
# read in salary csv
salary_df=pd.read_csv("raw-data/player_salary.csv")
salary_df
```

```python
# check data types
salary_df.dtypes
```

```python
# drop null values
salary_df.dropna(inplace=True)
```

```python
# drop Unnamed:0 & Tm columns
salary_columns = ["Player","season17_18"]
salary_clean_df = salary_df[salary_columns].copy()
salary_clean_df
```

```python
# rename columns
salary_final_df = salary_clean_df.rename(columns={"Player": "player_name",
                                                  "season17_18": "salary"
                                                 })
salary_final_df
```

```python
# drop duplicate salaries - traded in season players
salary_final_df.drop_duplicates(subset=['player_name'])
```

```python
# export clean salary csv
salary_final_df.to_csv("clean-data/salary_clean.csv",index=True, header = True)
```

Clean Stats Data

```python
# read in stats csv
stats_df = pd.read_csv("raw-data/player_stats.csv")
stats_df
```

```python
# check data types
stats_df.dtypes
```

```python
# drop null columns 
stats_df.dropna(inplace=True)
```

```python
# split player name column 
stats_df[['player_name','player_name2']]=stats_df['Player'].str.split("/",expand=True)
stats_df
```

```python
# select only necessary columns
stats_columns = ["player_name","Pos","Age", "Tm","G","MP","FG%","3P%","2P%","eFG%","FT%","TRB","AST","STL","BLK","PTS"]
stats_clean_df = stats_df[stats_columns].copy()
stats_clean_df
```

```python
# remove duplicate players
stats_grouped = stats_clean_df.groupby(['player_name'])
stats_final_df = stats_grouped.mean()
stats_final_df
```

```python
# rename columns
stats_final_df = stats_final_df.rename(columns={
                                                  "Age":"age",
                                                  "G":"games_played",
                                                  "MP": "minutes_played",
                                                  "FG%": "field_goal_percent",
                                                  "3P%": "three_point_percent",
                                                  "2P%": "two_point_percent",
                                                  "FT%":"free_throw_percent",
                                                  "eFG%":"effective_field_goal",
                                                  "TRB": "total_rebounds",
                                                  "AST": "assist",
                                                  "STL": "steal",
                                                  "BLK": "blocking",
                                                  "PTS": "points"
                                                 })
stats_final_df
```

```python
# round to two decimal places
stats_final_df.round(2)
```

```python
# export clean salary csv
stats_final_df.to_csv("clean-data/stats_clean.csv",index=True, header = True)
```

Create Extra Merged Dataframes

Stats + Salary df

```python
# merge stats and salary on player_name
stats_salary_df=pd.merge(salary_final_df, stats_final_df, on="player_name")
stats_salary_df
```

```python
stats_salary_df.index.name="index"
stats_salary_df
```

```python
# export stats/salary csv
stats_salary_df.to_csv("clean-data/stats_and_salary.csv",index=True, header = True)
```
